```python
from Bio import SeqIO
vic = [seq.seq for seq in SeqIO.parse("vic-pb2.fasta", "fasta")]

def get_seqs(variants_file):
    """
    variant_file: fasta file of PB2 sequences
    
    Returns: list of SeqRecords for all PB2 sequences in fasta file
    """
    all_seq = [seq for seq in SeqIO.parse(variants_file, "fasta")]
    return all_seq
def generate_tiles(get_seqs, tile_length, overlap, vic):
    """
    tile_length (int): the length of variable DNA per fragment
    overlap (int): length of overlap between tile fragments
    vic (str): PB2 sequence to use for overlap
    
    Returns: dictionary where keys are the id of the original PB2 sequence 
                and values are lists of the tiles generated from those sequences
    """
    tiles = {}
    for seq in get_seqs:
        seq_tiles = []
        for i in range(0, len(seq.seq)-1, tile_length+overlap):
            try:
                seq_tiles.append(vic[i-overlap:i] + seq.seq[i:i+tile_length] + vic[i+tile_length:i+tile_length+overlap])
            except:
                try:
                    seq_tiles.append(seq.seq[i:i+tile_length] + vic[i+tile_length:i+tile_length+overlap])
                except:
                    seq_tiles.append(vic[i-overlap:i] + seq.seq[i:i+tile_length])  
        tiles[seq.id] = seq_tiles
    return tiles

print(generate_tiles(get_seqs("AllPB2-2011-2016.fasta"), 60, 10, "vic-pb2.fasta"))
```
